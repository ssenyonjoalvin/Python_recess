```python
import pandas as pd

df = pd.read_csv("heart.csv")
print(df.to_string())
```

```python
# to check for empty cells/ missing values
print(df.isna().sum())
```

```python
print(df.duplicated().sum())
```

```python
#dropped all the duplicated rows 
# looking for repeated rows, values
df.drop_duplicates(inplace=True)
print(df.duplicated().sum())
```

```python
#explanatory data 
print(df.describe())
```

```python
# to find out the correlation between the  
df.corr()
```

```python
# data visualasition
import matplotlib.pyplot as plt
# line plot
plt.figure(figsize=(10, 6))
plt.plot(df['age'], df['thalach'], marker='o')
plt.title('Age vs. Maximum Heart Rate')
plt.xlabel('Age')
plt.ylabel('Maximum Heart Rate (thalach)')
plt.grid(True)
plt.show()
```

```python
# bar graph
plt.figure(figsize=(10, 6))
plt.bar(df['age'], df['chol'], color='skyblue')
plt.title('Age vs. Cholesterol Levels')
plt.xlabel('Age')
plt.ylabel('Cholesterol Levels (chol)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['age'], df['trestbps'], color='red')
plt.title('Age vs. Resting Blood Pressure')
plt.xlabel('Age')
plt.ylabel('Resting Blood Pressure (trestbps)')
plt.show()
```

```python
# histogram
plt.figure(figsize=(10, 6))
plt.hist(df['age'], color='green', edgecolor='black')
plt.title('Distribution of Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Seaborn visualisation
# bar plot
import seaborn as sns
plt.figure(figsize=(10, 5))
sns.barplot(x='age', y='chol', data=df, palette='viridis')
plt.title('Age vs. Cholesterol Levels')
plt.show()
```

```python
# scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='trestbps', data=df, color='red')
plt.title('Age vs. Resting Blood Pressure')
plt.show()
```

```python
# ploty Visualisation
import plotly.express as px

fig = px.line(df, x='age', y='thalach', title='Age vs. Maximum Heart Rate', markers=True)
fig.show()
```

```python
# bar plot
fig = px.bar(df, x='age', y='chol', title='Age vs. Cholesterol Levels', color='chol')
fig.show()

```
