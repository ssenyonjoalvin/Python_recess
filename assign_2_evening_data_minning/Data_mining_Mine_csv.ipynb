```python
import pandas as pd 
df = pd.read_csv("Mine.csv")
print(df.to_string)
```

```python
Missing_values_in_each_column = df.isna().sum()
print(Missing_values_in_each_column  )

duplicates = df.duplicated()
print(duplicates)

```

```python
# To hold empty cells in the date culomn we fill the empty cells with the date tha appears most
date_mode = df['Date'].mode()[0]
# print(data_mode)
df.fillna({'Date' : date_mode } , inplace=True)

# print(df.isna().sum())
print(df)



```

```python
# to ensure the there are no wrong formate in the data ,we convert all the data in the date culomn to the same formate

# df['Date'] = pd.to_datetime(df['Date'], format='%Y/%m/%d', errors='coerce')
print(df)
```

```python
# replacing all the missing value in the pulse column with the mean value
mean_pulse = df['Pulse'].mean()
df.fillna({'Pulse' : mean_pulse}, inplace=True)
print(df)
# print(df.isna().sum())
```

```python
# filling the null values with the mean 
mean_maxpulse = df['Maxpulse'].mean()
df.fillna({'Maxpulse':mean_maxpulse}, inplace=True)
print(df)
```

```python
# filling the Calories column with the median value of the column 
median_calories = df['Calories'].median()
df.fillna({'Calories':median_calories}, inplace=True)
print(df)
```
